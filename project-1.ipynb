```python
import pandas as pd
import pickle

# Sample data with lists for single row DataFrame
sample_data = pd.DataFrame({
    'Location': ['Rural'],
    'Cuisine': ['Mexican'],
    'Rating': [4.1],
    'Seating Capacity': [81],
    'Average Meal Price': [353.2],
    'Social Media Followers': [40081],
    'Chef Experience Years': [16],
    'Number of Reviews': [449],
    'Parking Availability': ['No'],
    'Weekend Reservations': [2],
    'Weekday Reservations': [1]
})

# Encoding 'Cuisine'
cuisine_encoding = {'Italian': 0.5, 'American': 0.4, 'Indian': 0.6, 'Mexican': 0.2, 'French': 0.3, 'Japan': 0.1}
sample_data['Cuisine'] = sample_data['Cuisine'].map(cuisine_encoding)


# Encoding 'Location'
location_encoding = {'Rural': 0, 'Urban': 2, 'SemiUrban': 1}
sample_data['Location'] = sample_data['Location'].map(location_encoding)

# Encoding 'Parking Availability'
parking_encoding = {'Yes': 1, 'No': 0}
sample_data['Parking Availability'] = sample_data['Parking Availability'].map(parking_encoding)

# Load the saved model from file
with open('random_forest_model.pkl', 'rb') as f:
    loaded_model = pickle.load(f)

# Ensure the sample data columns match the model's expected input
sample_data = sample_data[['Location', 'Cuisine', 'Rating', 'Seating Capacity', 
                           'Average Meal Price', 'Social Media Followers', 'Chef Experience Years', 
                           'Number of Reviews', 'Parking Availability', 'Weekend Reservations', 'Weekday Reservations']]

# Make predictions using the loaded model and sample data
predictions = loaded_model.predict(sample_data)

print("Predictions:", predictions)
```
